```python
import pandas as pd
import numpy as np
import math
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/AnjulaMehto/Sampling_Assignment/main/Creditcard_data.csv')
df
```

```python
print((df['Class']==0).sum())
print((df['Class']==1).sum())
print(df.shape)
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
sm = SMOTE(random_state=42)
x1, y1 = sm.fit_resample(x, y)
```

```python
print((y1==0).sum())
print((y1==1).sum())
```

```python
def models(df1):
  
  X_train, X_test, y_train, y_test = train_test_split(
    df1.iloc[:,0:-1], df1.iloc[:,-1], test_size=0.1, random_state=42)
  acc=[]

  #decision tree
  clf = tree.DecisionTreeClassifier()
  clf.fit(X_train, y_train)
  pdt=clf.predict(X_test)
  z=accuracy_score(y_test, pdt)
  acc.append(z)

  #random forest
  clf1 = RandomForestClassifier(n_estimators = 100) 
  clf1.fit(X_train,y_train)
  pdr=clf1.predict(X_test)
  z=accuracy_score(y_test,pdr)
  acc.append(z)

  #svm
  clf2 = svm.SVC(kernel='linear')
  clf2.fit(X_train,y_train)
  pds=clf2.predict(X_test)
  z=accuracy_score(y_test, pds)
  acc.append(z)

  #knn
  clf3 = KNeighborsClassifier(n_neighbors = 15)
  clf3.fit(X_train,y_train)
  pdk=clf3.predict(X_test)
  z=accuracy_score(y_test,pdk)
  acc.append(z)

  #lr
  clf4 = LogisticRegression(random_state=0,max_iter=1000)
  clf4.fit(X_train,y_train)
  pdl=clf4.predict(X_test)
  z=accuracy_score(y_test, pdl)
  acc.append(z)

  return acc
```

```python
ndf=pd.DataFrame(x1)
ndf['Class']=y1
ndf
```

```python
np.random.seed(0)
sample_size = 308
random_sample = ndf.sample(n=sample_size, random_state=0)
a1=models(random_sample.copy())
print(a1)
```

```python
n = len(ndf)
k = int(math.sqrt(n))
sample = ndf.iloc[::k]
a2=models(sample.copy())
print(a2)
```

```python
ssdf=ndf.groupby('Class', group_keys=False).apply(lambda x: x.sample(n=383))
a3=models(ssdf.copy())
print(a3)
```

```python
clusters = np.random.choice(np.arange(0,2), size=2, replace=False)
cluster_sample = ndf[ndf['Class'].isin(clusters)]
a4=models(cluster_sample.copy())
print(a4)
```

```python
dict={
    "random sampling":a1,
    "systematic sampling":a2,
    "stratified sampling":a3,
    "cluster sampling":a4
}

fdf=pd.DataFrame(dict,index=["Decision Tree","Random Forest","SVM", "KNN",
                             "Logistic Regression"])
fdf
```
